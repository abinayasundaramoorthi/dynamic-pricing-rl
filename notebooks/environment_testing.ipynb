## Test 1: `reset()`

Create the environment and reset it. It should return the starting state `[inventory, days_left]`.

```python
env = PricingEnv(initial_inventory=100, total_days=30)

state = env.reset()
print("Initial state:", state)

assert state == [100, 30], "reset() did not return the expected initial state!"
print("reset() PASSED")
```

## Test 2: `step()`

Take a single pricing action and check that the environment returns a valid next_state, reward, done flag, and info dict.

```python
action = 2  # pick a price level index (e.g. index 2 in price_levels list)

next_state, reward, done, info = env.step(action)

print("Next state:", next_state)
print("Reward:", reward)
print("Done:", done)
print("Info:", info)

assert isinstance(next_state, list) and len(next_state) == 2, "next_state should be [inventory, days_left]"
assert isinstance(reward, (int, float)), "reward should be a number"
assert isinstance(done, bool), "done should be a boolean"
print("step() PASSED")
```

## Test 3: `render()`

Check that render() prints the current state without errors.

```python
env.render()
print("render() PASSED")
```

## Test 4: Run Multiple Episodes with Random Actions

Run several full episodes, choosing random price actions each step, and log observations to confirm the environment runs without runtime errors.

```python
import random

NUM_EPISODES = 5
log = []

for episode in range(1, NUM_EPISODES + 1):
    state = env.reset()
    done = False
    total_reward = 0
    steps = 0

    while not done:
        action = random.randint(0, len(env.price_levels) - 1)
        next_state, reward, done, info = env.step(action)

        log.append({
            "episode": episode,
            "step": steps,
            "price": info["price"],
            "units_sold": info["units_sold"],
            "inventory_left": next_state[0],
            "days_left": next_state[1],
            "reward": reward
        })

        total_reward += reward
        steps += 1

    print(f"Episode {episode}: finished in {steps} steps | Total revenue: {total_reward:.2f}")

print("\nAll episodes completed with no runtime errors.")
```

## Test 5: Inspect Logged Observations

```python
import pandas as pd

log_df = pd.DataFrame(log)
log_df.head(10)
```
